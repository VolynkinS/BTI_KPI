```python
def create_forms(indicators, initial_dicts, entries, *, post=None):
    forms = {}
    for indicator, x, init_entry in zip(indicators, range(len(indicators)), initial_dicts):
        if indicator in [entry.indicator for entry in entries]:
            forms[indicator.id] = UserAnswerIndicatorForm(post, prefix=str(x),
                                                          instance=entries.get(indicator=indicator))
        else:
            forms[indicator.id] = UserAnswerIndicatorForm(post, prefix=str(x), initial=init_entry)
    return forms

criteria_on_page = Criteria.objects.filter(kpi__pk=1).prefetch_related('indicator_set', 'indicator_set__category')

indicators_on_page = Indicator.objects.filter(criteria__kpi__pk=1)
indicators_len = indicators_on_page.count()

initial_dicts = []
for indicator in indicators_on_page:
    d = {'indicator': indicator.pk, 'point': 0}
    initial_dicts.append(d)

entries_user = UserAnswerIndicator.objects.filter(user_answer__user=1, indicator__in=indicators_on_page)
# entries_list_indicator = [entry.indicator for entry in entries_user]


forms = {}
if entries_user.exists():
    for indicator, x, init_entry in zip(indicators_on_page, range(indicators_len), initial_dicts):
        if indicator in [entry.indicator for entry in entries_user]:
            forms[indicator.id] = UserAnswerIndicatorForm(prefix=str(x),
                                                          instance=entries_user.get(indicator=indicator))
        else:
            forms[indicator.id] = UserAnswerIndicatorForm(prefix=str(x))
else:
    for indicator, x, init_entry in zip(indicators_on_page, range(indicators_len), initial_dicts):
        forms[indicator.id] = UserAnswerIndicatorForm(prefix=str(x), initial=init_entry)

print(len(forms))
for x, y in forms.items():
    print(y.initial)
```
